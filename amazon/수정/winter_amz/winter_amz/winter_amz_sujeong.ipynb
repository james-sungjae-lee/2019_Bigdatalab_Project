````python
import json 

with open('items.json') as data_file:
    data = json.load(data_file)

len(data) # 현재 4502 아이템이 존재한다. 
#print(data[4501]['title']=='')
unable_id = []

for i in range(len(data)):
    # 전체 데이터에 대해서 data가 비어있으면, title이 비어있으면, 그 id를 가져와라.
    if data[i]['title'] == '': 
        unable_id.append(data[i]['id'])

print(len(unable_id))
````
