얘를 사용해서 해야하나?

현재 상태 - s3 에 winteramzimg라는 버켓이 있고 그 안에 폴더로 example이 들어가있고 그 안에 dataasins.json이라는 파일이 들어가있음 
```
--- winteramzimg
  |
  --- example
      |
      --- dataasins.json
```

그러면 이제 jupyter notebook에서 그 json 파일안에 들어있는 애들을 가지고 오도록 해보자!
- 으아악; 저렇게 안에 들어있는 폴더에서 가지고 오는 방법은 모르겠음
- 그냥 버켓 자체를 따로 쓰는것이 좋을듯

````python
import json 
import pandas

s3 = boto3.resource('s3')

# content_object를 접근할때 bucket안의 bucket에는 접근이 안된다. 에러가 남.
# 그러므로 이미지를 올릴때 하나하나 bucket을 나눠담아주는것이 좋겠다. 
# fashion_hoodie 한개, dress한개, shoes한개 이렇게 나눠야할듯

content_object = s3.Object('winteramzimg', 'dataasins.json') 
file_content = content_object.get()['Body'].read().decode('utf-8')
json_content = json.loads(file_content) # 이게 가지고오는것인듯. list 타입임

# 이렇게 하면 전부 다 긁어올 수 있음.
for i in range(len(json_content)):
    print(json_content[i]['product_data_asin'])

# 이게 되는걸 확인했으니까 연지의 파일을 가져다가 실행해보기로 해요~! 
````

````python
s3 = boto3.resource('s3')
bucket = s3.Bucket('winteramzimg')

json_list = []

for obj in bucket.objects.all(): # 근데 이렇게 하면 루프 겁나 많이 돌긴 하겟다,,
    content_object = s3.Object('winteramzimg', obj.key)
    file_content = content_object.get()['Body'].read().decode('utf-8')
    json_content = json.loads(file_content)
    json_list.append(json_content['features'])

#type of json_content['features'] 는 list 이다.
#전부 합친 리스트를 만든다.
print(json_list)
````
