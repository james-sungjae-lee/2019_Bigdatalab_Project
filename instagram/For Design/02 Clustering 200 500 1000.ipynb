```python
import pandas as pd
import numpy as np

file_name = 'allseason_20k_rawdata-Copy1.json'

df = pd.read_json(file_name)
```

```python
base_df = df[['id','hashtags']]
```

```python
base_df.index = base_df['id']
```

```python
base_df = base_df.drop(columns=['id'])
```

```python
fashion_df = base_df[base_df.astype(str)['hashtags'] != '[]']
```

```python
fashion_df.info()
```

```python
def getTagsDict(df):
    tag_dict = {}
    
    for index, row in df.iterrows():
        for tag in row['hashtags']:
            if tag in tag_dict:
                tag_dict[tag] = tag_dict[tag] + 1
            else:
                tag_dict[tag] = 1
                
    sort_dict = sorted(tag_dict.items(), key=lambda kv: kv[1])
    return sort_dict
```

```python
def get_DataFrame_for_affinity(df,taglist):
    
    result_df = pd.DataFrame(
        0, index = np.arange(0,len(df)),columns=taglist)    
    
    for i in range(len(df)):
        list_of_row = list(df.iloc[i])[0]
        
        for tag in list_of_row:
            if tag in taglist:
                result_df[tag][i] = 1
        
    return result_df
```

# Affinity

```python
tag_df = list(fashion_df['hashtags'])
for i in range(len(tag_df)):
    tag_df[i] = list(set(tag_df[i]))
    
fashion_df['hashtags'] = tag_df

fashion_dict = getTagsDict(fashion_df)
fashion_dict_top200 = fashion_dict[-200:]
fashion_top200 = []
for i in fashion_dict_top200:
    fashion_top200.append(i[0])
```

```python
fashion_dict_top500 = fashion_dict[-500:]
fashion_top500 = []
for i in fashion_dict_top500:
    fashion_top500.append(i[0])
```

```python
fashion_dict_top1000 = fashion_dict[-1000:]
fashion_top1000 = []
for i in fashion_dict_top1000:
    fashion_top1000.append(i[0])
```

```python
fashion_top200
```

```python
before_corr200 = get_DataFrame_for_affinity(fashion_df,fashion_top200)
```

```python
before_corr500 = get_DataFrame_for_affinity(fashion_df,fashion_top500)
```

```python
before_corr1000 = get_DataFrame_for_affinity(fashion_df,fashion_top1000)
```

```python
corr_affinity200 = before_corr200.corr(method = 'pearson')
```

```python
corr_affinity500 = before_corr500.corr(method = 'pearson')
```

```python
corr_affinity1000 = before_corr1000.corr(method = 'pearson')
```

```python
corr_threshold200 = corr_affinity200.copy()
corr_threshold500 = corr_affinity500.copy()
corr_threshold1000 = corr_affinity1000.copy()
```

```python
def addone_dividetwo(x):
    return pow((x+1)/2, 2)
```

```python
corr_threshold200 = corr_threshold200.applymap(addone_dividetwo)
```

```python
corr_threshold500 = corr_threshold500.applymap(addone_dividetwo)
```

```python
corr_threshold1000 = corr_threshold1000.applymap(addone_dividetwo)
```

# Clustering

```python
from sklearn.cluster import SpectralClustering
```

```python
sc_200 = SpectralClustering(n_clusters=14,affinity='precomputed')
sc_500 = SpectralClustering(n_clusters=30,affinity='precomputed')
sc_1000 = SpectralClustering(n_clusters=50,affinity='precomputed')
```

```python
predict200 = sc_200.fit_predict(corr_threshold200)
predict500 = sc_500.fit_predict(corr_threshold500)
predict1000 = sc_1000.fit_predict(corr_threshold1000)
```

```python
predict200 = pd.DataFrame(predict200)
predict500 = pd.DataFrame(predict500)
predict1000 = pd.DataFrame(predict1000)

predict200.columns = ['Group']
predict500.columns = ['Group']
predict1000.columns = ['Group']
```
