```python
def getTagsDict(df):
    tag_dict = {}
    
    for index, row in df.iterrows():
        for tag in row['hashtags']:
            if tag in tag_dict:
                tag_dict[tag] = tag_dict[tag] + 1
            else:
                tag_dict[tag] = 1
                
    sort_dict = sorted(tag_dict.items(), key=lambda kv: kv[1])
    return sort_dict
```

```python
tag_df = list(fashion_df['hashtags'])
for i in range(len(tag_df)):
    tag_df[i] = list(set(tag_df[i]))
    
fashion_df['hashtags'] = tag_df
```

# Affinity

```python
def get_DataFrame_for_affinity(df,taglist):
    
    result_df = pd.DataFrame(
        0, index = np.arange(0,len(df)),columns=taglist)    
    
    for i in range(len(df)):
        list_of_row = list(df.iloc[i])[0]
        
        for tag in list_of_row:
            if tag in taglist:
                result_df[tag][i] = 1
        
    return result_df
```

```python
def getTaglist(fashion_dict):
    tag_list = []
    
    for i in fashion_dict:
        tag_list.append(i[0])
        
    return tag_list
```

# Silhouette

```python
from sklearn.cluster import SpectralClustering
from sklearn import metrics
```

```python
def getSScores(corr_threshold,start,end):
    X = np.array(corr_threshold)

    SScore_list = []

    for i in range(start,end):
        pearson_sc = SpectralClustering(n_clusters = i, affinity='precomputed')
        predict_sc = pearson_sc.fit_predict(corr_threshold)
        labels = predict_sc
        score = metrics.silhouette_score(X, labels, metric='euclidean')
#         print('n_cluster = {}'.format(i))
#         print(round(score,4))
#         print()
        SScore_list.append(round(score, 4))
    return SScore_list
```

```python
def getClusterOneHot(groupby_tags200,fashion_dict200,before_corr200):
    results = []

    for i in range(len(groupby_tags200)):
        cluster_list = groupby_tags200.iloc[i]
        cluster_one_hot = pd.DataFrame(before_corr200.loc[0])
        cluster_one_hot = cluster_one_hot.T

        for tag in fashion_dict200:
            cluster_one_hot[tag[0]] = 0

        for tag in cluster_list:
            cluster_one_hot[tag] = 1

        resultlist = []
        for index, row in before_corr200.iterrows():
            logical_and = row & cluster_one_hot
            resultlist.append(logical_and)

        sumlist = []
        for row in resultlist:
            sumvalue = row.sum(axis = 1)
            sumlist.append(sumvalue)

        resultpd = pd.DataFrame(sumlist, post_id)
        resultpd = resultpd.sort_values(by=0, ascending = False)
        resultpd.columns = ['Freq']

        topvalue = resultpd.iloc[0]['Freq']  # ex) 5
        topid_list = []
        for i in range(3):
            tmppd = resultpd[resultpd['Freq'] == topvalue - i]
            topid_list.append(tmppd.index.values)

        result = (cluster_list, topid_list,topvalue)
        results.append(result)
    
    return results
    
```

```python
def getResultDataFrame(results,post_id_username2):
    result_df = []
    for i in range(len(results)):
        cluster = results[i]
        cluster_list = cluster[0]
        cluster_idlist = cluster[1]
        topvalue = cluster[2]

        topclusters = []
        for j in range(len(cluster_idlist)):
            topcluster_df = pd.DataFrame(cluster_idlist[j])
            topcluster_df.columns = ['post_id']

            username_list = []
            for i in range(len(topcluster_df)):
                id = topcluster_df.iloc[i]['post_id']
                username_list.append(post_id_username2.loc[id]['username'][0])

            topcluster_df['username'] = username_list
            topcluster_df = pd.DataFrame(topcluster_df.groupby('username')['post_id'].apply(list))

            freq = np.empty(len(topcluster_df))
            freq.fill(topvalue)
            freq = freq.astype(int)

            topcluster_df['Freq'] = freq

            topclusters.append(topcluster_df)

        result_df.append(topclusters)
    return result_df

    
```
