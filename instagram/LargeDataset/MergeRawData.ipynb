```python
import json

file_list = []

for i in range(8700):
    if i == 0 :
        filename = 'fallfashion' + str(0) + 'to' + str(next_end) + '_rawdata.json'
    else:
        filename = 'fallfashion' + str(next_start) + 'to' + str(next_end) + '_rawdata.json'
        
    print(filename)
    
    with open(filename) as file:
        rawdata = file.read()
        rawdata = json.loads(rawdata)
        file_list.append(rawdata)
        file.close()
    
    next_start = next_end + 1
    next_end = next_end + size
    
    if next_end > 5010:
        break
    
```

```python
def save_json_file(my_tag, myjson):
    file_name = my_tag + '_rawdata.json'
    with open(file_name, 'w') as file:
        json.dump(myjson, file)
```
